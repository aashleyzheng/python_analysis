```python
import numpy as np

array = np.arange(12)
array.shape
```

```python
#adarray都是一維
#這些寫法都是改變shape的屬性

array1 = array.reshape((3, 4))
array1.shape
array1[1,2]
```

```python
array2 = array1.reshape((2, 6))
array2[1,0]
```

```python
array3 = array2.reshape((1, 2, 1, 6, 1))
array3[0,1,0,0,0]
```
